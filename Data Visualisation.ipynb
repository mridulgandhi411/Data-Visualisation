# Data Visualisation 

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
winered=pd.read_csv('./datasets/winequality-red.csv',sep=';')
winered.head()
```

```python
winered.info()
```

```python
winered.describe()
```

```python
winered['quality'].unique()
```

```python
plt.hist(winered['quality'])
```

```python
plt.hist(winered['quality'],bins=range(3,10,1))
```

```python
plt.hist(winered['quality'],bins=range(3,10,1),width=0.7)
```

```python
winered['quality'].plot(kind='hist',bins=range(3,10,1),width=0.7)
```

```python
winered['quality'].value_counts()
```

```python
winered['quality'].value_counts().plot(kind='bar')
```

```python
winered['quality'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
```

```python
# we want this sorted about the quality 
sorted_values=winered['quality'].value_counts().sort_index()
sorted_values.plot(kind='bar')
plt.xticks(rotation=0)
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.title('Wine Quality')
```

```python
plt.plot(winered['fixed acidity'])
```

```python
plt.plot(winered['fixed acidity'])
plt.xlim(0,100)
```

```python
plt.plot(winered['fixed acidity'])
plt.xlim(0,100)
plt.ylim(5,12)
```

```python
plt.plot(winered['fixed acidity'])
plt.plot(winered['pH'])
```

```python
# to get the box we use plt.legend()
plt.plot(winered['fixed acidity'],label='Fixed acidity')
plt.plot(winered['pH'],label='pH')
plt.legend()
```

```python
plt.plot(winered['fixed acidity'],label='Fixed acidity')
plt.plot(winered['pH'],label='pH')
plt.legend(loc='upper left')
```

```python
# if we want both in seperate graphs
plt.subplot(1,2,1)
#this means : represent it as  1 row 
#2 columns 
# 1 is the plot number(3rd argument)
plt.plot(winered['fixed acidity'])
plt.title('Fixed Acidity')
plt.xlabel('number of wines')
plt.subplot(1,2,2)
plt.plot(winered['pH'])
plt.title('pH')

#for common title
plt.suptitle('Acidity and pH')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(winered['fixed acidity'],color='green')
plt.title('Fixed Acidity')
plt.xlabel('number of wines')
plt.subplot(1,2,2)
plt.plot(winered['pH'],color='red')
plt.title('pH')

#for common title
plt.suptitle('Acidity and pH')
plt.show()
```

```python
# to expand the width of this graph we use figsize in plt.figure
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(winered['fixed acidity'],color='green')
plt.title('Fixed Acidity')
plt.xlabel('number of wines')
plt.subplot(1,2,2)
plt.plot(winered['pH'],color='red')
plt.title('pH')

#for common title
plt.suptitle('Acidity and pH')
plt.show()
```

```python
plt.scatter(x=winered['fixed acidity'],y=winered['pH'])
plt.xlabel('fixed acidity')
plt.ylabel('pH')
plt.show()
```

```python
plt.scatter(x=winered['fixed acidity'],y=winered['pH'],alpha=0.3)
plt.xlabel('fixed acidity')
plt.ylabel('pH')
plt.show()
```

```python
plt.scatter(x=winered['fixed acidity'],y=winered['pH'],alpha=0.3,color='green')
plt.xlabel('fixed acidity')
plt.ylabel('pH')
plt.show()
```

# There is a Library for making Much Better Plots 
# Seaborn

```python
import seaborn as sns
```

```python
sns.jointplot(x='fixed acidity',y='pH',data=winered)
plt.show()
```

```python
#lmplot-linear model plot
sns.lmplot(x='fixed acidity',y='pH',data=winered)
plt.show()
```

```python
sns.regplot(x='fixed acidity',y='pH',data=winered,order=1)
plt.show()
```

```python
sns.regplot(x='fixed acidity',y='pH',data=winered,order=2)
plt.show()
```

```python
sns.regplot(x='fixed acidity',y='pH',data=winered,order=3)
plt.show()
```

```python
sns.regplot(x='fixed acidity',y='pH',data=winered,order=6)
plt.show()
```
